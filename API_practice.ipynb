```python
#Function to return a json object from the nhtsa api
def get_api_data(url):
    r = requests.get(url)
    data = r.json()
    data = data['Results']
    return data
```

```python
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetVehicleTypesForMake/mercedes?format=json'
mercedes = get_api_data(url)
```

```python
print(hondas.head(10))
print(len(hondas))
```

```python
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/getmodelsformake/mercedes?format=json'
mercedes = get_api_data(url)
mercedes = pd.DataFrame(mercedes)
print(mercedes.head(10))
print(len(mercedes))
```

```python
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/getmodelsformake/toyota?format=json'
toyota = get_api_data(url)
toyota = pd.DataFrame(toyota)
print(toyota.head())
print(len(toyota))
```

```python
all_cars = pd.concat([hondas, mercedes, toyota])
print(all_cars.sample(10))
```

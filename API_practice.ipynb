```python
#Function to return a json object from the nhtsa api
def get_api_data(url):
    r = requests.get(url)
    data = r.json()
    data = data['Results']
    return data
```

```python
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetVehicleTypesForMake/mercedes?format=json'
mercedes = get_api_data(url)
```

```python
vin_numbers = ['JN8AZ1MW9EW500262;5FNRL5H61FB054227;1J4BA5H13BL619297;4M2DU55PXVUJ38891;JHMZE2H32CS006277']
```

```python
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/DecodeVINValuesBatch/'
post_fields = {'format': 'json', 'data':vin_numbers}
r = requests.post(url, data=post_fields)
r = r.json()
print(r)
```

```python
data = r['Results']
data = pd.DataFrame(data)
```

```python
data.head()
```

```python
data = data.transpose()
data.head()
```

```python
print(data.info())
```

```python
data.columns = (data.loc['Make'] + ' ' + data.loc['Model'])
```

```python
data.head()
```

```python
data.columns = (data.loc['Make'].str.capitalize() + ' ' + data.loc['Model'])
data.head()
```

```python
data.isnull().sum()
```

```python
data.replace('', np.NaN, inplace=True)
```

```python
print(data.isnull().sum())
print(len(data))
```

```python
data.dropna(axis=0, how='all', inplace = True)
data.head()
```
